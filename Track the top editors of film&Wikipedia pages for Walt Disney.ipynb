```
%sql
-- Find all the pages including the prev_titles
SELECT DISTINCT
      prev_title
  FROM tbl_refer
  INNER JOIN tbl_film
      ON curr_title = value
-- Seems like there isnt any referal pages that got editted 
```

```
%sql
select * from tbl_article
where title = 'Alexander and the Terrible, Horrible, No Good, Very Bad Day (film)'
-- It seesm like there is only 1 record in the given file vs 33 edits shown in this link
-- https://xtools.wmflabs.org/articleinfo/en.wikipedia.org/Alexander%20and%20the%20Terrible,%20Horrible,%20No%20Good,%20Very%20Bad%20Day%20%28film%29/2015-02-01/2015-02-28
```

```
# Add a column that represents the average time between edits for the user (null if they only have one edit)
spark_all_title = spark.sql('''
SELECT DISTINCT
      prev_title as value
  FROM tbl_refer
  INNER JOIN tbl_film
      ON curr_title = value
  UNION ALL
  select * from tbl_film
''')
spark_all_title.createOrReplaceTempView('tbl_all_titles')
```

```
%sql
select * from tbl_all_titles 
```

```
# To make it more convenient, maybe use the above table to get all the titles for ease of query
final_DF =  spark.sql("""SELECT
    revisionUsername as user_name,
    COUNT(revisionId) as edits
  FROM tbl_article AS a
  INNER JOIN tbl_all_titles AS b
    ON a.title = b.value
  GROUP BY user_name""")
final_DF.show()
```
